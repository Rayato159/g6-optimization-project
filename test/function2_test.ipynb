```python
import numpy as np
import pandas as pd
from matplotlib import pyplot as plt
from matplotlib import cm
```

```python
df = pd.read_csv("../graph/exam2_output.csv")
df.head()
```

```python
X = df[["x1", "x2"]]
z = df["z"]
```

```python
from sklearn.neural_network import MLPRegressor
```

```python
N = 100
```

```python
X1_new = np.linspace(-2, 2, 100)
X2_new = np.linspace(-2, 2, 100)
X_predict = np.array([X1_new, X2_new])
```

```python
NN = MLPRegressor(hidden_layer_sizes=([N]*7), activation="relu", random_state=1, max_iter=2000)
NN.fit(X, z)
predict = NN.predict(X_predict)
print(NN.score(X, z))
```

```python
X1 = df["x1"]
X2 = df["x2"]
Z = np.array(df['z'])
```

```python
fig, ax = plt.subplots(1, 2, figsize=(10,5))

ax[0].set_tx[0].tricontour(X1, X2, Z, 50, linewidths=0.5, colors='k')
ax[0].tricontourf(X1, X2, Z, 50, cmap=cm.jet)

ax[1].set_title("predict")
ax[1].tricontour(X1, X2, predict, 50, linewidths=0.5, colors='k')
ax[1].tricontourf(X1_new, X2_new, predict, 50 ,cmap=cm.jet)

plt.show()
```

```python
print(f"x1: {X1[predict.argmin()]} x2: {X2[predict.argmin()]} z: {predict[predict.argmin()]}")
```

```python
coef = NN.coefs_[0]
intercept = NN.intercepts_[0]
print(coef.shape)
print(intercept.shape)
```

```python
fig = plt.figure()
ax = fig.add_subplot(projection='3d')

sctt_true = ax.scatter3D(X1, X2, Z, c = Z, cmap = cm.jet, marker='o', alpha=0.3, label="true")
sctt_predict = ax.scatter3D(X1_new, X2_new, predict, c = Z, cmap = cm.hot, marker='^', label="predict")

plt.title("Exam2 3D scatter plot")

plt.legend
plt.grid()
plt.legend()
plt.show()
```
